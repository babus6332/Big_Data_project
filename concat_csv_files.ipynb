# Concat all csv files from archive csvYourBoardingData

```python
import pandas as pd
import glob

# Путь к директории с файлами csv извлеченными из архива
path = r'./data/csvYourBoardingData'
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
frame = frame.loc[:, ~df.columns.str.contains('^Unnamed')]
```
