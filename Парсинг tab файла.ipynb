```python
import pandas as pd
```

```python
with open ('Sirena-export-fixed.tab','r',encoding = 'utf-8') as f:
    lines = f.readlines()
    for count,i in enumerate(lines):
        while "  " in lines[count]:
            lines[count] = lines[count].replace("  ", " ")
```

```python
lines[0].split(' ')
```

```python
columns = ['PaxName','PaxBirthDate','DepartDate','DepartTime','ArrivalDate','ArrivalTime','FlightCodeSh','From','Dest',
           'Code_e_Ticket','TravelDoc','Seat','Meal','TrvClsFare','Baggage','PaxAdditionalInfo' ,'AgentInfo']
```

```python
PaxName = []
PaxBirthDate = []
DepartDate = []
DepartTime = []
ArrivalDate = []
ArrivalTime = []
FlightCodeSh = []
From = []
Dest = []
Code_e_Ticket = []
TravelDoc = []
Seat = []
Meal = []
TrvClsFare = []
Baggage = []
PaxAdditionalInfo = []
AgentInfo = []

for line in lines[1:]:
        b = line.split(' ')
        PaxName.append(' '.join(b[0:3]))
        PaxBirthDate.append(b[3])
        DepartDate.append(b[4])
        DepartTime.append(b[5])
        ArrivalDate.append(b[6])
        ArrivalTime.append(b[7])
        FlightCodeSh.append(b[8])
        From.append(b[9])
        Dest.append(b[10])
        Code_e_Ticket.append(b[11])
        TravelDoc.append(b[12]+b[13])
        Seat.append(b[14]) #b[14] везде N/A
```

```python
# Meal = []
# TrvClsFare = []
# Baggage = []
# AgentInfo = []
# PaxAdditionalInfo = []


for line in lines[1:]:
    b = line.split(' ')
    if(len(b[15]) == 1):
        Meal.append('N/A')
        TrvClsFare.append(b[15])
    elif(len(b[15]) > 2):
        Meal.append(b[15])
        if(len(b[16])==1):
            TrvClsFare.append(b[16])
        elif ((len(b[16]) == 6) or (len(b[16]) == 9)):
            TrvClsFare.append('N/A')
            Baggage.append(b[16])
        else:
            TrvClsFare.append('N/A')
            
    else:
        Meal.append('N/A')
            
    if ((len(b[16]) == 6) or (len(b[16]) == 9)):
        Baggage.append(b[16])
        
    elif ((len(b[17]) == 6) or (len(b[17]) == 9)):
            Baggage.append(b[17])
    else:
            Baggage.append('N/A')
            
    if((len(b[-2])>1)and(b[-2].isdigit()==False)):
        AgentInfo.append(b[-2])
    else:
        AgentInfo.append('N/A')
        PaxAdditionalInfo.append(b[-2])
        continue

    if((len(b[-4])!=6)or (len(b[-4])!=9)):
        if(b[-3].isdigit()==False):
            PaxAdditionalInfo.append('N/A')
        else:
            PaxAdditionalInfo.append(b[-4] + ' ' + b[-3])
    else:
        PaxAdditionalInfo.append('N/A')
```

```python
df = pd.DataFrame({'PaxName':PaxName,'PaxBirthDate':PaxBirthDate,'DepartDate':DepartDate,'DepartTime':DepartTime,
              'ArrivalDate':ArrivalDate,'ArrivalTime':ArrivalTime,'FlightCodeSh':FlightCodeSh,'From':From,
             'Dest':Dest,'Code_e_Ticket':Code_e_Ticket,'TravelDoc':TravelDoc,'Meal':Meal,
                  'TrvClsFare':TrvClsFare,'Baggage':Baggage,'PaxAdditionalInfo':PaxAdditionalInfo,'AgentInfo':AgentInfo},columns = columns)
df.head()
```

```python
df.to_csv('Excellent__Sirena-export-fixed.csv',index=False)
```
