```python
import pandas as pd
import numpy as np
import json
```

```python
with open("FrequentFlyerForum-Profiles.json") as f:
    jsn = json.load(f)
```

```python
isHeader = True
for profile in jsn['Forum Profiles']:
    df1 = pd.json_normalize(profile, record_path=['Registered Flights'], 
                           meta = ['Sex', 'NickName', ['Real Name', 'Last Name'], ['Real Name', 'First Name']]
                          )
    df2 = pd.json_normalize(profile, record_path=['Travel Documents'], meta = ['NickName'], meta_prefix=['Travel Documents'])
    
    df3 = pd.json_normalize(profile, record_path=['Loyality Programm'], meta = ['NickName'], meta_prefix=['Travel Documents'])
    
    df1 = pd.merge(df1, df2, on='NickName', how='left')
    df1 = pd.merge(df1, df3, on='NickName', how='left')
    if isHeader:
        df1.to_csv('FrequentFlyerForum-Profiles.csv', mode = 'w', index=False)
    else:
        df1.to_csv('FrequentFlyerForum-Profiles.csv', mode = 'a', header=False, index=False)
    isHeader = False
```
