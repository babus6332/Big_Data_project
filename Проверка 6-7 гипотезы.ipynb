```python
data2['DepartDate']=data2['DepartDate'].astype(str)+' '+data2['DepartTime'].astype(str)+':00'
data2['ArrivalDate']=data2['ArrivalDate'].astype(str)+' '+data2['ArrivalTime'].astype(str)+':00'
```

```python
data2['DepartDate'] = pd.to_datetime(data2['DepartDate'], format='%Y-%m-%d %H:%M:%S')
data2['ArrivalDate'] = pd.to_datetime(data2['ArrivalDate'], format='%Y-%m-%d %H:%M:%S')
data2.sort_values(['PaxName', 'PaxBirthDate', 'DepartDate'], inplace = True)
```

```python
data2['Next step']= data2.groupby('TravelDoc').shift(-1)['DepartDate']
#Используем для подсчета длительности
def delta(x):
    t0 = x.split('*^')[0]
    t1 = x.split('*^')[1]
    return((pd.to_datetime(t0) - pd.to_datetime(t1)).seconds)

data2['Длительность'] = data2['Next step'].astype(str) + '*^' + data2['ArrivalDate'].astype(str)
data2['Длительность'] = data2['Длительность'].apply(delta)
```

```python
data2['Next step'] = pd.to_datetime(data2['Next step'], format='%Y-%m-%d %H:%M:%S')
data2['ArrivalDate'] = pd.to_datetime(data2['ArrivalDate'], format='%Y-%m-%d %H:%M:%S')

data2['Разница'] = data2['Next step']-data2['ArrivalDate']
```
