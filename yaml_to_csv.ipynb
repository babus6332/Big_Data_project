```python
#output_dir = 'yaml_csv'
isHeader = True
for file in os.listdir('yaml-parsed/'):
    with open('yaml-parsed/'+file) as f:
        jsn = json.load(f)
        df = pd.json_normalize(jsn, record_path=['FF'], meta = ['DATA', 'FLIGHT', 'FROM', 'STATUS', 'TO'])
        df[['PROGRAMM','NUMBER']] = df['CARD'].str.split(' ',expand=True)
        df.drop('CARD', axis=1, inplace=True)
        if isHeader:
            df.to_csv("Skyteam_Exchange.csv", mode='w', index = False)
        else:
            df.to_csv("Skyteam_Exchange.csv", mode='a', header = False, index = False)
        isHeader = False
        print("file {} done, shape = {}".format(file, df.shape))
```
